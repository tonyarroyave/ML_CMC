```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from os import listdir
from os.path import isfile, join
from pandas.plotting import radviz
```

```python
def open_file(name, path):
    file_path = path + name
    return pd.read_csv(file_path)

def get_avrg(df_data):
    df_data['Open*'] = pd.to_numeric(df_data['Open*'], errors='coerce').fillna(0.0)
    df_data['Close**'] = pd.to_numeric(df_data['Close**'], errors='coerce').fillna(0.0)
    return (df_data['Open*']+df_data['Close**'])/2

def create_time_col(path):
    file_path = path + 'bitcoin.csv'
    btc_df = pd.read_csv(file_path)
    return pd.to_datetime(btc_df['Date'])
```

```python
plt.figure(figsize=(14,6))
the_path = os.getcwd()
data_path = the_path + '/data/'
onlyfiles = [f for f in listdir(data_path) if isfile(join(data_path, f))]
target = 'bitcoin'
```

```python
count = 0
big_df = pd.DataFrame()
for file in onlyfiles:
    count += 1
    df_data = open_file(file, data_path)
    col_avrg = get_avrg(df_data)
    if len(col_avrg) == 183:
        big_df[file.split('.')[0]] = col_avrg
big_df.head(5)
```

```python
big_df['Date'] = create_time_col(data_path)
big_df.set_index('Date', inplace=True)
big_df.head(5)
```

```python
"""Linear correlation analysis"""
correlations = big_df.corr()[target].sort_values(ascending=False)
```

```python
"""Getting the coins with higher correlations"""
keys_h_corr = correlations.keys()
keys = []
vals = []
for curr_corr in keys_h_corr:
    if abs(correlations[curr_corr]) > 0.70 and abs(correlations[curr_corr]) != 1.0:
        keys.append(curr_corr)
        vals.append(correlations[curr_corr])
h_corr = pd.Series(vals, index=keys)
```
